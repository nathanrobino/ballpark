# Bellman Stages Decomposition

This notebook follows the model presented in section four of "Regional Consumption Responses and the Aggregate Fiscal Multiplier" {cite:t}`Dupor2023-bn`. Here we abstract from the behavior of the intermediate firms and regional mutual funds in each region, treating them as exogenous to the households. 

This document shows how to decompose one period of the household problem into a sequence of modular stages. 

## Households in Region i

Households maximize lifetime utility by choosing consumption and labor supply. 

\begin{equation*}
U=\mathbf{E}_{0}\left[\sum_{t=0}^{\infty} \beta^{t-1}\left\{\log \left(c_{i, t}\right)+\psi \frac{\left(1-h_{i, t}\right)^{1-\theta}}{1-\theta}\right\}\right],
\end{equation*}

They are subject to idiosyncratic productivity shocks, receive rate of return on assets saved in a regional mutual fund (risky return), and dividends from intermediate good firms that depend on the average productivity of workers in the region. 

## Bellman Form (from Original)

\begin{align*}
V_{t}\left(x_{t}, a_{t}, \beta ; \phi_{i, t}\right)= & \max _{c_{t}, a_{t+1}, h_{t}}\left\{\log \left(c_{t}\right)+\psi \frac{\left(1-h_{t}\right)^{1-\theta}}{1-\theta}\right. \\
& \left.+\beta \sum_{x_{t+1}} \Gamma_{x_{t}, x_{t+1}} V_{t+1}\left(x_{t+1}, a_{t+1}, \beta ; \phi_{i, t+1}\right)\right\} \\
\text { s.t. } \quad & c_{t}+a_{t+1}=w_{i, t} x_{t} h_{t}-\mathcal{T}(.)+\left(1+r_{i, t}\right) a_{t}+(1-\omega) \delta\left(x_{t}\right) D_{i, t},  \\
& a_{t+1} \geq 0 .
\end{align*}

where $c$ denotes consumption, $h$ denotes work, $\psi$ is a parameter governing utility of leisure, $\theta$ is the Frisch elasticity of labour supply, $\beta$ is the idiosyncratic discount factor, $\Gamma$ is a transition matrix approximating the productivity shock process $x$, $a$ denotes assets saved in a regional mutual fund, $w$ denotes wages, $\mathcal{T}$ is the tax schedule, and $\delta\left(x\right) D$ is the share of dividends received by households.

### `shocks-only` Stage

Instead of the approximate process $\Gamma_{x_t,x_{t+1}}$, consider the full AR(1) process: 

\begin{equation*}
\log x_{t+1}=\rho \log x_{t}+\eta_{t+1} \quad \text { with } \eta_{t+1} \sim \operatorname{iid} N\left(0, \sigma_{\eta}^{2}\right)
\end{equation*}

Using the notation from the original paper, this stage has perches

Perch        | Indicator | State      | value functions                    | Explanation 
------------ | --------- | ---------- | ----------------------------       | ------------
Arrival      |  $\prec$  | $\bar{x}$  | $v_{\prec} = E_{\prec}[v_{\succ}]$ | pre-shock value
Decision     |  $\sim$   | $\bar{x}$  | (none)                             | no choice
Continuation |  $\succ$  | $x$    | $v_\succ$                          | post-shock value

### `labor-supply` stage

Given the worker's productivity, worker chooses labor supply to maximize their leisure and continuation value function.

Let $b = w_{i, t} x_{t} h_{t}-\mathcal{T}(.)+\left(1+r_{i, t}\right) a_{t}+(1-\omega) \delta\left(x_{t}\right) D_{i, t}$. If we treat the aggregate wage and dividend as parameters exogenous to the hosuehold, then once $h$ is chosen, this b is determined exactly. Thus the labor supply decision transforms $x$ and $a$ into $b$.

Perch        | Indicator | State                   | value functions              | Explanation 
------------ | --------- | ----------------------- | ---------------------------- | ------------
Arrival      |  $\prec$  | $(x,a)$  | $v_{\prec} = v_{\sim}$       | shocks considered in separate stage
Decision     |  $\sim$   | $(x,a)$  | $v_{\sim}(x,a) = \max_h E[\psi \frac{\left(1-h_{t}\right)^{1-\theta}}{1-\theta} + v_{\succ}(b(h_{t}))]$ | choose labor supply
Continuation |  $\succ$  | $b$        | $v_{\succ}$                    | value at exit

### consumption (`cons-noshocks`) stage

Perch        | Indicator | State  | value functions                          | Explanation 
------------ | --------- | ------ | ---------------------------------------- | ------------
Arrival      |  $\prec$  | $b$    | $v_{\prec} = v_{\sim}$                   | no pre-decision shocks
Decision     |  $\sim$   | $b$    | $v_{\sim}(m) = \max_c log(c) + v(b − c)$ | choose consumption
Continuation |  $\succ$  | $a$    | $v_\succ$                                | post-shock value

### Productivity Connector

In order to restart the next period, we need to reclassify $x$ as $\bar{x}$. This would be done with a connector. 

## All Stages

Element        | Transition | Action

`cons-noshocks`     |  $b \implies a$  |  choose consumption
