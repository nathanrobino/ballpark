## Why in-ballpark for Econ-ARK

The paper develops a partial-equilibrium life-cycle overlapping-generations model with uninsurable idiosyncratic income risk, aggregate interest-rate and house-price risk, and fixed-rate mortgages, in which heterogeneous households make a discrete choice among renting, owning without adjusting, and owning while adjusting their housing stock and mortgage, followed by a continuous choice over non-durable consumption, short-term savings, and mortgage balance, subject to a loan-to-value constraint and a fixed adjustment cost.

The model ties a robust empirical fact — young households' consumption is two-to-three times more responsive to monetary-policy-induced interest-rate shocks than older households' — to age-varying mortgage-refinancing propensities, and uses this structural link to quantify how population aging dampens the aggregate transmission of monetary policy.

It is in-ballpark for Econ-ARK because it is a canonical heterogeneous-agent life-cycle problem with discrete-continuous choice, an explicit CRRA-Cobb-Douglas kernel, a warm-glow bequest, and both idiosyncratic and aggregate shocks — exactly the class of modular-DP problems Econ-ARK targets for formalization.
