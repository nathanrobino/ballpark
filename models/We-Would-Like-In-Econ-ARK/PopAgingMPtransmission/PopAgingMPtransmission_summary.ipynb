# Paper Summary

## Prior Literature and Motivation

Demographics and monetary policy transmission had been studied in largely separate literatures. Work on demographics ({cite:t}`Auerbach1984-qt,Auerbach1989-nb`; {cite:t}`Rios-Rull1996-jp,Rios-Rull2001-fa`) analyzed how aging affects capital accumulation and asset prices, while work on transmission ({cite:t}`Kaplan2014-ee`; {cite:t}`Auclert2019-dr`; {cite:t}`Mian2013-wr`) showed how household balance-sheet heterogeneity shapes consumption responses to interest rates. A small literature ({cite:t}`Imam2015-ee`; {cite:t}`Juselius2015-dk`) had documented correlations between aging and weaker monetary transmission, but without structural foundations. {cite:t}`Wong2021-sr` was among the first to explicitly link population aging to monetary policy effectiveness through age-varying consumption responses.

*For a more detailed discussion of the prior literature, see [Prior Literature](PopAgingMPtransmission_prior-literature.ipynb).*

## Overview

* **Research Question: What are the effects of demographic changes on the transmission of monetary policy to consumption?**
* **Part I: Empirical estimate**
    * **Age-specific consumption elasticities to interest rate**
        * **Consumption of young people is two-three times more responsive to interest rate shocks**
        * **Consumption elasticities decline with age**
    * **Consumption response driven by homeowners**
        * **refinance or new loans following interest rate decline**
* **Part II: Life-cycle model - partial equilibrium overlapping generations economy**
    * **Uninsurable labor income risk, life-cycle saving motive, fixed-rate mortgage structure**
    * **Fixed costs to adjust long-term assets**
        * **housing and fixed-rate mortgage**

## The Model

The economy is populated by a continuum of overlapping-generations households indexed by $j$, each living for a maximum of $T$ periods. A household aged $a$ survives to the next period with probability $\pi_a$; households work for the first $T_y$ periods and retire thereafter. Preferences are time-separable with discount factor $\beta$.

### Preferences and bequest

Per-period flow utility over non-durable consumption $c_{jat}$ and housing services $h_{jat}$ is Cobb-Douglas within CRRA:

$$
u(c_{jat}, h_{jat}) = \frac{\left(c_{jat}^{\alpha} \, h_{jat}^{1-\alpha}\right)^{1-\sigma} - 1}{1 - \sigma}.
$$

Upon death, total net wealth $W_{jat}$ is bequeathed and evaluated under a warm-glow CRRA term with the Cocco-Gomes-Maenhout shift:

$$
V^{\mathrm{beq}}(W_{jat}) = B \cdot \frac{\left(W_{jat} - 1\right)^{1-\sigma}}{1-\sigma},
$$

where $B$ is the bequest utility weight (calibrated as in {cite:t}`Cocco2005-ub`).

### Discrete-continuous choice

Each working-age household first chooses among **four** regimes — rent, own and do not refinance the existing mortgage, own and refinance the existing mortgage, or purchase a new home — and then, conditional on the regime, chooses non-durable consumption, short-term savings, and (in the refinance and purchase regimes) housing stock and mortgage balance:

$$
V_{j a t}(z_{j a t}) = \max \left\{
V_{j a t}^{\mathrm{rent}},\;
V_{j a t}^{\mathrm{own\,\&\,no\text{-}refi}},\;
V_{j a t}^{\mathrm{own\,\&\,refi}},\;
V_{j a t}^{\mathrm{purchase}}
\right\}\!(z_{j a t}).
$$

The state vector decomposes into age, aggregate state, labor income, and a four-vector of arrival-period household assets:

$$
z_{j a t} = \bigl(\, a,\; S_{t},\; y_{j a t},\; \mathrm{assets}_{j, a-1, t-1} \,\bigr),
\qquad
\mathrm{assets}_{j, a-1, t-1} = \bigl(\, s_{j, a-1, t-1},\; h_{j, a-1, t-1}^{\mathrm{own}},\; b_{j, a-1, t-1},\; R_{j, a-1, t-1} \,\bigr),
$$

where $s$ is one-period short-term assets, $h^{\mathrm{own}}$ is the housing stock owned at arrival (depreciating at rate $\delta$), $b$ is the outstanding mortgage balance, and $R$ is the household-specific fixed coupon on the existing mortgage. $R$ is a *household state*, not a market object: its recursion is defined in "Mortgage dynamics" below.

### Shock processes

There are two independent sources of uncertainty.

**Idiosyncratic income.** Log labor income decomposes into a deterministic age profile $\chi_a$, a persistent idiosyncratic component $\eta_{jat}$, and an age-specific loading $\phi_a$ on the aggregate income level:

$$
\log y_{jat} = \chi_{a} + \eta_{jat} + \phi_{a}\!\left(y_{t}\right),
\qquad
\eta_{jat} = \rho_\eta \, \eta_{j, a-1, t-1} + \psi_{jt},
\qquad
\psi_{jt} \overset{\mathrm{iid}}{\sim} \mathcal{N}\!\left(0, \sigma_\eta^2\right).
$$

**Aggregate state.** The aggregate vector $S_t$ — containing log aggregate income $\log y_t$, log house prices $\log p_t$, and the one-period real interest rate $r_t$ — follows a first-order VAR

$$
S_{t} = A_{0} + A_{1} \, S_{t-1} + u_{t},
\qquad
u_{t} \sim \mathcal{N}\!\left(0, V\right),
$$

with $u_t$ multinormal and mean-zero. The paper discretizes $S_t$ onto an 18-state Markov chain via the Tauchen method for computation; the continuous VAR is kept in this statement but the discretization is load-bearing for the formalization layer (see "Conventions and scope" below).

### Timing within a period

Each period $t$ unfolds as:

1. The household arrives at age $a$ carrying $(s_{j, a-1, t-1},\; h^{\mathrm{own}}_{j, a-1, t-1},\; b_{j, a-1, t-1},\; R_{j, a-1, t-1},\; \eta_{j, a-1, t-1})$ and observes the lagged aggregate state $S_{t-1}$.
2. The aggregate innovation $u_t$ is drawn and $S_t = A_0 + A_1 S_{t-1} + u_t$ is realized. The duration-specific mortgage rate $r_t^{d}$ and the rental rate $p_t^r$ are determined by $S_t$ through the reduced-form functions $f^d(\cdot)$ and $f^{pr}(\cdot)$ below.
3. The idiosyncratic innovation $\psi_{jt}$ is drawn; $\eta_{jat}$ and labor income $y_{jat}$ are realized.
4. The household chooses a regime $\in \{\mathrm{rent},\, \mathrm{no\text{-}refi},\, \mathrm{refi},\, \mathrm{purchase}\}$, then the continuous controls conditional on that regime.
5. End-of-period assets $(s_{j a t},\, h^{\mathrm{own}}_{j a t},\, b_{j a t},\, R_{j a t})$ are set; the mortgage balance and coupon evolve per the laws of motion in the next section; survival $\pi_a$ resolves.
6. Continuation: with probability $\pi_a$ the household enters period $t+1$ at age $a+1$; with probability $1-\pi_a$ it dies and $V^{\mathrm{beq}}$ applies.

### Mortgage dynamics

Following {cite:t}`Campbell2003-qk` and Kaplan, Moll, and Violante (2018), the mortgage is amortized over the remaining life of the agent. A new loan originated at age $a$ in period $t$ has duration $d(a) = T - a$ and is priced at the current market rate on a $d(a)$-duration mortgage, $R_{j a 0} = r_{t}^{d(a)}$. The corresponding mortgage payment $M_{j a t}$ is the level annuity implied by the outstanding balance, coupon, and remaining duration:

$$
b_{j a t} = M_{j a t} \sum_{k=1}^{d(a)} \frac{1}{\left(1 + R_{j a t}\right)^{k}}
\qquad\Longrightarrow\qquad
M_{j a t} = b_{j a t} \left[\sum_{k=1}^{d(a)} \frac{1}{\left(1 + R_{j a t}\right)^{k}}\right]^{-1}.
$$

$M_{j a t}$ is therefore a **mechanical deduction** from $(b_{j a t},\, R_{j a t},\, d(a))$, not an additional control.

The balance and coupon evolve state-contingently on the regime choice. Let $\mathbf{1}(\mathrm{adj})_{j a t} = 1$ if the household chooses the `refi` or `purchase` regime at period $t$ (both reset the coupon) and $0$ under `rent` or `no-refi`. Then

$$
R_{j, a+1, t+1} = r_{t+1}^{d(a+1)}\, \mathbf{1}(\mathrm{adj})_{j, a+1, t+1} + R_{j a t}\,\bigl[1 - \mathbf{1}(\mathrm{adj})_{j, a+1, t+1}\bigr],
$$

and the balance evolves as
$b_{j, a+1, t+1} = b_{j a t}(1 + R_{j a t}) - M_{j a t}$ under `no-refi`;
$b_{j, a+1, t+1} = b_{j a t}$ (a fresh choice) under `refi` and `purchase`;
and $b_{j, a+1, t+1} = 0$ under `rent` (the existing loan is paid off in full via the $b_{j, a-1, t-1}(1 + R_{j, a-1, t-1})$ term in the rent-regime budget).

Two reduced-form links carry the aggregate state into household-facing prices:

$$
r_{t}^{d} = f^{d}(S_t) = a_{0}^{d} + a_{1}^{d}\, r_{t} + a_{2}^{d}\, \log y_{t},
\qquad
\log p_{t}^{r} = f^{pr}(S_t) = \alpha_{0} + \alpha_{1}\, r_{t} + \alpha_{2}\, \log y_{t} + \alpha_{3}\, \log p_{t},
$$

with $f^{d}$ estimated for $d \in \{1,\, 15,\, 30\}$-year mortgages and interpolated across intermediate durations, and $f^{pr}$ estimated on national house-price and rent indices. These reduced-form mappings are what carry monetary-policy shocks (innovations to $r_{t}$) into the mortgage rate the household faces and into the rental cost of housing.

A minimum equity requirement
$$
b_{j a t} \leq (1 - \phi)\, p_{t}\, h^{\mathrm{own}}_{j a t}
$$
binds in the `refi` and `purchase` regimes; the short-term borrowing constraint $s_{j a t} \geq -\underline{s}$ applies in every regime.

Transaction costs differ by regime:

- **`purchase`**: $F_{j a t}^{\mathrm{new}} = \kappa^{\mathrm{new}}\, p_{t}\, h^{\mathrm{own}}_{j a t}$, proportional to the value of the newly purchased home (calibrated $\kappa^{\mathrm{new}} = 0.05$).
- **`refi`**: $F_{j a t}^{\mathrm{refi}} = F_{f}^{\mathrm{refi}} + F_{v}^{\mathrm{refi}}\, b_{j a t}$, a dominant fixed component $F_{f}^{\mathrm{refi}}$ (calibrated to ≈ \$2{,}100) plus a small variable component $F_{v}^{\mathrm{refi}}$ (set to zero in the baseline).

### 1. Rent regime

The household rents, liquidates any arrival-period owned housing stock at the depreciated value, and pays off the existing mortgage (if any) in full. The value function is

$$
V_{j a t}^{\mathrm{rent}}\!\left(z_{j a t}\right)
= \max_{c_{j a t},\, h_{j a t}^{\mathrm{rent}},\, s_{j a t}}
\; \frac{\left(c_{j a t}^{\alpha}\, \bigl(h_{j a t}^{\mathrm{rent}}\bigr)^{1-\alpha}\right)^{1-\sigma} - 1}{1 - \sigma}
+ E_{j a t}\!\left[\tilde V_{j, a+1, t+1}\!\left(z_{j, a+1, t+1}\right)\right],
$$

subject to

$$
\begin{aligned}
c_{j a t} + s_{j a t} + p_{t}^{r}\, h_{j a t}^{\mathrm{rent}}
&= y_{j a t} + (1-\delta)\, p_{t}\, h_{j, a-1, t-1}^{\mathrm{own}} + (1 + r_{t})\, s_{j, a-1, t-1} - b_{j, a-1, t-1}\!\left(1 + R_{j, a-1, t-1}\right), \\
h_{j a t}^{\mathrm{own}} &= 0, \qquad b_{j a t} = 0, \qquad s_{j a t} \geq -\underline{s}.
\end{aligned}
$$

The continuation-value operator $\tilde V$ is the survival-weighted combination of the within-model continuation and the bequest, defined in "Retirement and death" below. The coupon state $R_{j a t}$ is undefined in this regime because no mortgage is outstanding at the end of the period.

### 2. Own & no-refi regime

The household keeps the existing home, makes the scheduled mortgage payment $M_{j a t}$ defined in "Mortgage dynamics" above, and does *not* pay the fixed refinancing cost. Housing services are flowed from the depreciated stock. The value function is

$$
V_{j a t}^{\mathrm{own\,\&\,no\text{-}refi}}\!\left(z_{j a t}\right)
= \max_{c_{j a t},\, s_{j a t}}
\; \frac{\left(c_{j a t}^{\alpha}\, \bigl[(1-\delta)\, h_{j, a-1, t-1}^{\mathrm{own}}\bigr]^{1-\alpha}\right)^{1-\sigma} - 1}{1 - \sigma}
+ E_{j a t}\!\left[\tilde V_{j, a+1, t+1}\!\left(z_{j, a+1, t+1}\right)\right],
$$

subject to

$$
\begin{aligned}
c_{j a t} + s_{j a t} &= y_{j a t} + (1 + r_{t})\, s_{j, a-1, t-1} - M_{j a t}, \\
h_{j a t}^{\mathrm{own}} &= (1-\delta)\, h_{j, a-1, t-1}^{\mathrm{own}}, \\
b_{j a t} &= b_{j, a-1, t-1}\!\left(1 + R_{j, a-1, t-1}\right) - M_{j a t}, \\
R_{j a t} &= R_{j, a-1, t-1}, \qquad s_{j a t} \geq -\underline{s}.
\end{aligned}
$$

The coupon $R_{j a t}$ carries over from arrival; the residual balance $b_{j a t}$ updates by the amortization identity. $M_{j a t}$ is the mechanical annuity implied by $(b_{j, a-1, t-1},\, R_{j, a-1, t-1},\, d(a))$ and is not an additional control.

### 3. Purchase regime

The household sells the depreciated arrival-period home at $(1-\delta)\, p_{t}\, h^{\mathrm{own}}_{j, a-1, t-1}$, pays off the existing mortgage at $b_{j, a-1, t-1}(1 + R_{j, a-1, t-1})$, buys a new home $h_{j a t}^{\mathrm{own}}$ at price $p_{t}$, originates a new mortgage $b_{j a t}$ at the current market rate $r_{t}^{d(a)}$, and pays the proportional transaction cost $F_{j a t}^{\mathrm{new}} = \kappa^{\mathrm{new}}\, p_{t}\, h_{j a t}^{\mathrm{own}}$. The value function is

$$
V_{j a t}^{\mathrm{purchase}}\!\left(z_{j a t}\right)
= \max_{c_{j a t},\, s_{j a t},\, h_{j a t}^{\mathrm{own}},\, b_{j a t}}
\; \frac{\left(c_{j a t}^{\alpha}\, \bigl(h_{j a t}^{\mathrm{own}}\bigr)^{1-\alpha}\right)^{1-\sigma} - 1}{1 - \sigma}
+ E_{j a t}\!\left[\tilde V_{j, a+1, t+1}\!\left(z_{j, a+1, t+1}\right)\right],
$$

subject to

$$
\begin{aligned}
c_{j a t} + s_{j a t} + p_{t}\, h_{j a t}^{\mathrm{own}} - b_{j a t} + F_{j a t}^{\mathrm{new}}
&= y_{j a t} + (1-\delta)\, p_{t}\, h_{j, a-1, t-1}^{\mathrm{own}} + (1 + r_{t})\, s_{j, a-1, t-1} - b_{j, a-1, t-1}\!\left(1 + R_{j, a-1, t-1}\right), \\
b_{j a t} &\leq (1-\phi)\, p_{t}\, h_{j a t}^{\mathrm{own}}, \\
R_{j a t} &= r_{t}^{d(a)}, \qquad s_{j a t} \geq -\underline{s}, \\
F_{j a t}^{\mathrm{new}} &= \kappa^{\mathrm{new}}\, p_{t}\, h_{j a t}^{\mathrm{own}} \quad (\kappa^{\mathrm{new}} = 0.05\ \text{in the baseline calibration}).
\end{aligned}
$$

The coupon resets to the current market rate on a $d(a)$-duration mortgage; the housing stock is freshly chosen; the transaction cost is proportional to the purchased value, making $F_{j a t}^{\mathrm{new}}$ endogenous to the chosen $h_{j a t}^{\mathrm{own}}$.

### 4. Own & refi regime

The household keeps the existing (depreciated) home but refinances the mortgage: the old balance $b_{j, a-1, t-1}(1 + R_{j, a-1, t-1})$ is paid off, a new mortgage $b_{j a t}$ is originated at the current market rate $r_{t}^{d(a)}$, and the refinancing cost $F_{j a t}^{\mathrm{refi}}$ is paid. The value function is

$$
V_{j a t}^{\mathrm{own\,\&\,refi}}\!\left(z_{j a t}\right)
= \max_{c_{j a t},\, s_{j a t},\, b_{j a t}}
\; \frac{\left(c_{j a t}^{\alpha}\, \bigl[(1-\delta)\, h_{j, a-1, t-1}^{\mathrm{own}}\bigr]^{1-\alpha}\right)^{1-\sigma} - 1}{1 - \sigma}
+ E_{j a t}\!\left[\tilde V_{j, a+1, t+1}\!\left(z_{j, a+1, t+1}\right)\right],
$$

subject to

$$
\begin{aligned}
c_{j a t} + s_{j a t} - b_{j a t} + F_{j a t}^{\mathrm{refi}}
&= y_{j a t} + (1 + r_{t})\, s_{j, a-1, t-1} - b_{j, a-1, t-1}\!\left(1 + R_{j, a-1, t-1}\right), \\
h_{j a t}^{\mathrm{own}} &= (1-\delta)\, h_{j, a-1, t-1}^{\mathrm{own}}, \\
b_{j a t} &\leq (1-\phi)\, p_{t}\, h_{j a t}^{\mathrm{own}}, \\
R_{j a t} &= r_{t}^{d(a)}, \qquad s_{j a t} \geq -\underline{s}, \\
F_{j a t}^{\mathrm{refi}} &= F_{f}^{\mathrm{refi}} + F_{v}^{\mathrm{refi}}\, b_{j a t}.
\end{aligned}
$$

The coupon resets to the current market rate (as in `purchase`), but the housing stock is carried through at its depreciated value and no proportional home-value cost is paid. The refinancing cost is predominantly fixed ($F_{f}^{\mathrm{refi}}$ calibrated to ≈ \$2{,}100); the paper sets the small variable component $F_{v}^{\mathrm{refi}} = 0$ in its baseline calibration.

### Retirement and death

From age $T_y + 1$ onward the household is retired and receives a deterministic Social Security transfer that is a function of its income in the final working-age period (modelled as in {cite:t}`Guvenen2014-lo`). The recursive problem otherwise retains the four-regime structure above, with the stochastic labor-income process replaced by this retirement transfer.

Upon death, the household's total net wealth (liquidated housing plus short-term assets net of outstanding mortgage) is

$$
W_{j a t} = (1-\delta)\, p_{t}\, h_{j, a-1, t-1}^{\mathrm{own}} + (1 + r_{t})\, s_{j, a-1, t-1} - b_{j, a-1, t-1}\!\left(1 + R_{j, a-1, t-1}\right),
$$

evaluated under the shifted warm-glow bequest introduced in "Preferences and bequest" above:

$$
V^{\mathrm{beq}}(W_{j a t}) = B \cdot \frac{\left(W_{j a t} - 1\right)^{1-\sigma}}{1-\sigma}.
$$

The age-$a$ continuation value in each regime is the survival-weighted combination of the within-model continuation and the bequest terminal value:

$$
E_{j a t}\!\left[\, \tilde V_{j, a+1, t+1} \,\right]
= \pi_{a}\, E_{j a t}\!\left[\, V_{j, a+1, t+1}\!\left(z_{j, a+1, t+1}\right) \,\right]
+ \left(1 - \pi_{a}\right)\, V^{\mathrm{beq}}\!\left(W_{j, a+1, t+1}\right).
$$

Substituting $E_{j a t}[\tilde V_{j, a+1, t+1}]$ for $E_{j a t}[V_{j, a+1, t+1}]$ in each of the four regime Bellman equations yields the full recursive problem; at the terminal age $T$ the continuation reduces to $V^{\mathrm{beq}}$.

### Conventions and scope

Three modelling choices pin down how the four-regime recursive problem above translates into a formalized stage, and are worth stating explicitly because they are easy to misread from the paper alone.

**Free regime toggling.** A household that arrives owning housing is not locked into the owner side: the `rent` regime is available every period and is entered by liquidating the arrival-period stock at its depreciated value and paying off any outstanding mortgage (Eq. 12 in the paper). Conversely, a renter arriving with $h^{\mathrm{own}}_{j, a-1, t-1} = b_{j, a-1, t-1} = 0$ can enter the `purchase` regime next period. There is no hysteresis; the four regimes form a flat discrete-choice menu in every period.

**$M_{j a t}$ is a mechanical deduction, not a state.** The payment $M_{j a t}$ is the level annuity implied by $(b_{j, a-1, t-1},\, R_{j, a-1, t-1},\, d(a))$ in the `no-refi` regime, and is not carried as an extra state variable — the balance recursion $b_{j, a+1, t+1} = b_{j a t}(1 + R_{j a t}) - M_{j a t}$ is a deterministic identity. In dolo-plus terms this is a **mechanical (non-optimized) deduction** — it should be flagged as such when the YAML draft is produced, per CONTRIBUTING.md §"Formalization layer".

**Aggregate state is a finite Markov chain in practice.** The paper writes the aggregate block as a Gaussian VAR (Eq. 8) but discretizes it with Tauchen onto an 18-state Markov chain for computation. The duration-specific mortgage rate $r_{t}^{d} = f^{d}(S_t)$ and the rental rate $p_{t}^{r} = \exp f^{pr}(S_t)$ are tabulated on that grid. A formalizer who treats the aggregate block as continuous Gaussian innovations will over-parameterize the shock structure relative to what the paper actually computes; the intended target for dolo-plus is the Markov-chain form.

## Results

* **The covariance mechanism.** Young households are simultaneously more likely to refinance when rates fall (larger loans and longer durations mean interest savings exceed the fixed refinancing cost) and more likely to be at their short-term liquidity constraint (they borrow against high expected future income). The *positive covariance* between refinancing propensity and constrained-MPC status is what amplifies the aggregate consumption response to monetary-policy shocks, and is what erodes as the population ages.
* **The refinancing and new-lending channel drives the age gradient.** The model matches the empirical fact that the consumption response of the young to interest-rate shocks is two-to-three times that of the old, and attributes this gap to the fact that young homeowners are disproportionately the ones who refinance or originate new loans after expansionary shocks.
* **Fixed- vs. variable-rate counterfactual.** Replacing the FRM structure with an ARM structure (no fixed costs, universal pass-through of the short rate to mortgage payments) produces two offsetting effects: the covariance between refinancing and constrained-MPC status collapses (dampening cross-sectional heterogeneity and the aggregate response), while more households see their payments change (amplifying the aggregate response). Quantitatively the second dominates: aggregate consumption responds *more* under ARMs, while the young-old cross-sectional gap is *smaller*.
* **Quantitative decomposition.** Shutting down refinancing and re-estimating under ARMs reduces the young-old consumption-response gap by approximately 45%, i.e. the refinancing channel accounts for roughly 45% of the observed age gradient. The remaining share is attributable to income and house-price channels that themselves vary with age.
* **Aggregate implication.** Holding the mortgage structure fixed, shifting the population age distribution toward older households lowers the aggregate covariance term and therefore dampens the aggregate consumption response to a monetary-policy shock — population aging is, through this channel, a force that reduces monetary-policy effectiveness.

## Subsequent Literature and Research Directions

The literature following {cite:t}`Wong2021-sr` has deepened understanding of the refinancing channel as a state- and region-dependent mechanism ({cite:t}`statedependent_eichenbaum_2022`; {cite:t}`regional_beraja_2019`; {cite:t}`mortgage_berger_2021`), extended the framework into heterogeneous-agent New Keynesian (HANK) models ({cite:t}`monetary_kaplan_2016`), and examined how mortgage market design and distributional consequences shape transmission ({cite:t}`mortgage_guren_2018`; {cite:t}`innocent_coibion_2017`).
