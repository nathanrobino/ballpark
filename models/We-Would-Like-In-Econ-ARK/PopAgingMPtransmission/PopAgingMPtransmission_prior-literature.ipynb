# Prior Literature

## Where the paper sits in the literature

The paper integrates three prior literatures that had largely developed in isolation.

### Connecting demographics to monetary policy transmission

Prior work had treated these as separate topics. The demographic-macroeconomics tradition — exemplified by the life-cycle general-equilibrium models of {cite:t}`Rios-Rull2001-fa` — studied how age structure shapes capital accumulation and asset prices without taking a stance on monetary policy. The household-heterogeneity and monetary-policy tradition — exemplified by {cite:t}`Kaplan2014-ee` and {cite:t}`Auclert2019-dr` — studied how balance-sheet heterogeneity shapes consumption responses to interest rates without taking a stance on age structure. The paper is among the first to link the two, asking whether an aging population systematically changes how monetary policy works.

### Integrating multiple transmission channels

The literature had identified several channels through which monetary policy affects consumption — housing wealth effects, mortgage refinancing, and redistribution between borrowers and savers. {cite:t}`Mian2013-wr` established, at regional scale, that household balance sheets drive consumption responses. {cite:t}`Di-Maggio2014-rv` documented the refinancing channel directly, showing that households who refinance to lower rates raise consumption substantially. The paper's contribution is to show how these channels vary systematically with age, because younger households are more likely to be borrowers carrying mortgages while older households are more likely to hold interest-bearing assets.

### Policy-relevant implications for aging societies

{cite:t}`Imam2015-ee` had begun exploring whether demographics weaken monetary policy, but relied on reduced-form correlations. The paper provides a structural, life-cycle micro-foundation — modeling how life-cycle consumption behavior interacts with monetary transmission — rather than simply documenting correlations.

## Key foundational papers

- **{cite:t}`Kaplan2014-ee`**, *A Model of the Consumption Response to Fiscal Stimulus Payments* — introduced the "wealthy hand-to-mouth" taxonomy and demonstrated that balance-sheet heterogeneity, and in particular the distinction between liquid and illiquid assets, is crucial for understanding how policy affects consumption. This framework is central to understanding why age — which correlates strongly with asset composition — matters for transmission.

- **{cite:t}`Auclert2019-dr`**, *Monetary Policy and the Redistribution Channel* — decomposed monetary transmission into income, wealth, and interest-rate-exposure channels and showed how household heterogeneity determines aggregate policy effectiveness. This framework directly motivates asking how demographic composition affects transmission.

- **{cite:t}`Mian2013-wr`**, *Household Balance Sheets, Consumption, and the Economic Slump* — demonstrated empirically at the regional level the powerful connection between household balance sheets, housing wealth, and consumption during the Great Recession, providing both methodological tools and motivation for studying how household characteristics mediate policy effects.
