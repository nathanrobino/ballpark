# Subsequent Literature

## Research directions that emerged

The literature citing {cite:t}`Wong2021-sr` has developed along several interconnected trajectories.

### 1. The refinancing channel as a core transmission mechanism

A major strand has focused intensively on mortgage refinancing as a key channel through which monetary policy affects household consumption. This includes empirical work documenting the magnitude of these effects ({cite:t}`interest_dimaggio_2017`; {cite:t}`mortgage_abel_2021`; {cite:t}`refinancing_amromin_2020`) and theoretical work modeling the state-dependent nature of refinancing behavior ({cite:t}`statedependent_eichenbaum_2022`; {cite:t}`mortgage_berger_2021`).

### 2. Heterogeneous-agent New Keynesian (HANK) models

The field has moved decisively toward incorporating household heterogeneity into macroeconomic models of monetary policy. {cite:t}`monetary_kaplan_2016` is a watershed moment, establishing that heterogeneous-agent models produce fundamentally different predictions about monetary transmission than representative-agent models. This framework has become the new benchmark for studying monetary policy.

### 3. Regional and state-dependent heterogeneity

Researchers have explored how the effects of monetary policy vary across space and time. {cite:t}`regional_beraja_2019` shows that the refinancing channel's strength depends on the regional distribution of housing equity, while {cite:t}`statedependent_eichenbaum_2022` demonstrates that monetary policy effectiveness depends on the state of the economy and the history of interest rates.

### 4. Housing market structure and mortgage design

{cite:t}`mortgage_guren_2018` and {cite:t}`mortgage_greenwald_2016` examine how institutional features of mortgage markets — contract terms, prepayment options, down-payment requirements — shape monetary transmission. This connects to policy debates about optimal mortgage market design.

### 5. Distributional consequences of monetary policy

{cite:t}`innocent_coibion_2017` and the broader literature on monetary policy and inequality examine how monetary policy affects different households differently, with implications for both efficiency and equity.

---

## What's cutting-edge now

The frontier of this literature:

- **Path dependence**: how past interest rates and refinancing decisions affect current policy effectiveness ({cite:t}`mortgage_berger_2021`).
- **Endogenous housing and mortgage choices**: models where households' tenure and mortgage decisions respond to policy, creating feedback effects ({cite:t}`effects_kinnerud_2025`; {cite:t}`mortgage_guren_2018`).
- **Quantitative HANK models**: calibrated models that match micro-level consumption and balance-sheet data while generating realistic aggregate dynamics.
- **Interaction of multiple channels**: how refinancing, wealth effects, income effects, and redistribution channels interact and potentially offset each other.

## Most important papers for understanding where the field is heading

### 1. {cite:t}`statedependent_eichenbaum_2022` — *State-Dependent Effects of Monetary Policy: The Refinancing Channel*

Co-authored by Wong herself, this is a direct extension of the 2016 work. It shows that monetary policy's effectiveness varies substantially depending on the distribution of existing mortgage rates in the economy. When many households carry mortgages at rates above current market rates, monetary easing is powerful; when most have already refinanced, the channel is muted. This state dependence has implications for how central banks should think about policy in different environments and connects to debates about forward guidance and quantitative easing.

### 2. {cite:t}`regional_beraja_2019` — *Regional Heterogeneity and the Refinancing Channel of Monetary Policy*

Shows empirically that the refinancing channel's strength depends critically on local housing market conditions. Regions where homeowners had negative or low equity could not refinance even when rates fell, weakening monetary transmission precisely where stimulus was most needed. This geographic heterogeneity has implications for understanding why monetary policy may have limited effectiveness following housing busts and connects to policy discussions about programs like HARP.

### 3. {cite:t}`mortgage_berger_2021` — *Mortgage Prepayment and Path-Dependent Effects of Monetary Policy*

Develops a structural model showing how the history of interest rates affects the current distribution of mortgage rates, creating path dependence in monetary policy effectiveness. It provides a rigorous framework for understanding why the refinancing channel varies over time and offers tools for quantifying how past policy decisions constrain current policy options. This represents the cutting edge of integrating mortgage market microstructure into macroeconomic models.
